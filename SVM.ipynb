```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
iris.feature_names
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel('sepal length (cms)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], color = 'red', marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], color= 'blue', marker='x')
# plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'], color= 'green', marker='o')
```

```python
plt.xlabel('petal length (cms)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], color = 'red', marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color= 'blue', marker='x')
```
