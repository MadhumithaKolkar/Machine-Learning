```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income($US)')
plt.scatter(df.year,df['per capita income (US$)'], color = "green", marker = "X")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income($US)')
plt.scatter(df['year'], df['per capita income (US$)'], color='green', marker='X')

# Plotting the linear regression line giving the year and getting the prediction of per capita income for that year
plt.plot(df['year'], reg.predict(df[['year']]), color='blue')
```

```python
df.to_csv('Canada_per_capita_prediction.csv',index=False)
```
